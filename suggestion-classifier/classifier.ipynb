```python
modal = ['should be ','could be ','would be ','might be ','may be ','will be ','should do ','could do ','would do ','might do ','may do ','can do ','will do ','should have ','could have ','would have ','might have ','may have ','can have ','will have ','needs to ','need to ','should come ','could come ','might come ','wish ','hope ','hopefully ','Allow']
```

```python
sentences = []
for i in range(0,len(csv)):
    sentences.append(csv['text'][i])
```

```python
is_suggestion = []
for i in range(0,len(sentences)):
    flag = 0
    for word in modal:
        if word in sentences[i]:
            flag = 1
            break
    if flag == 1:
        is_suggestion.append(1)
    else:
        is_suggestion.append(0)
```

```python
out = pd.Series(is_suggestion)
```

```python
from sklearn.metrics import accuracy_score
print accuracy_score(out,csv['class'])
```

```python
def get_features(Sentence):
    words = Sentence.split(' ')
    phone_names = ['moto','samsung','oppo','vivo','lenovo','coolpad','redmi']
    features = {}

    features['do'] = ('would do' or 'could do' or 'should do' or 'might do') in Sentence
    #'would be' or 
    features['be'] = ('could be' or 'should be' or 'might be') in Sentence
    features['have'] = ('would have' or 'could have' or 'should have' or 'might have') in Sentence
    features['come'] = ('would come' or 'could come' or 'should come' or 'might come') in Sentence
    features['maycome'] = ('will come' or 'may come') in Sentence
    features['maydo'] = ('will do' or 'may do') in Sentence
    features['mayhave'] = ('will have' or 'may have') in Sentence
    features['maybe'] = ('will be' or 'may be') in Sentence
    features['need'] = ('need to' or 'needs to') in Sentence
    features['wish'] = ('wish' or 'hope' or 'hopefully') in Sentence
    #features['please'] = ('Please' or 'please' or 'Pls' or 'pls') in Sentence
    features['allow'] = ('Allow' or 'allow') in Sentence
    
    return features
```

```python
featurinks = [(get_features(n),v) for n,v in zip(csv['text'],csv['class'])]
```

```python
import nltk
def get_features(Sentence):
    words = Sentence.split(' ')
    verbs = ['VB','VBD','VBG','VBN','VBP','VBZ']
    flag = False
    if 'should' in words :
        totag = [words[words.index('should')+1]]
        flag = nltk.pos_tag(totag) in verbs
    features = {}

    features['do'] = ('would do' or 'could do' or 'should do' or 'might do') in Sentence
    features['be'] = ('could be' or 'should be' or 'might be') in Sentence 
    features['have'] = ('would be' or 'would have' or 'could have' or 'should have' or 'might have' or 'should\'ve') in Sentence
    features['come'] = ('would come' or 'could come' or 'should come' or 'might come') in Sentence
    features['maycome'] = ('will come' or 'may come') in Sentence
    features['maydo'] = ('will do' or 'may do') in Sentence
    features['mayhave'] = ('will have' or 'may have') in Sentence
    features['maybe'] = ('will be' or 'may be') in Sentence
    features['need'] = ('need to' or 'needs to') in Sentence
    features['wish'] = ('wish' or 'hope' or 'hopefully') in Sentence
    features['please'] = ('Please' or 'please' or 'Pls' or 'pls' or 'Dear') in Sentence
    features['allow'] = ('Allow' or 'allow') in Sentence
    features['want'] = ('wants to' or 'want to') in Sentence
    features['must'] = ('must be' or 'must do' or 'must have') in Sentence
    features['verb'] = flag
    
    
    return features
```
